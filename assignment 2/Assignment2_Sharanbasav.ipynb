## Creating Database and Inserting Data 

```python
import sqlite3

# Connecting to the sqlite3 database
dbpath = 'mydb.sqlite'
conn = sqlite3.connect(dbpath)
```

```python
cur = conn.cursor()
```

```python
drop_stories_query  = 'DROP TABLE IF EXISTS stories;'
drop_info_query = 'DROP TABLE IF EXISTS word_count;'

#Now we run both queries
cur.execute(drop_stories_query)
cur.execute(drop_info_query)
```

```python
cur.executescript("""
CREATE TABLE stories(
    story_id INTEGER PRIMARY KEY,
    words TEXT 
);
""")
```

```python
cur.executescript("""
CREATE TABLE word_count(
    word_count_id INTEGER PRIMARY KEY,
    words TEXT,
    count INTEGER,
    story_id INTEGER,
    FOREIGN KEY(story_id) REFERENCES stories(story_id));
""")
```

```python
stories = ["story-1.txt", "story-2.txt","story-3.txt", "story-4.txt" ]

for story in stories:
    words = []
    count_of_each_word = {}
    story_id = ""
    try:      
        with open( story, encoding='utf-8') as fp:            
            story_text = fp.read()          
            insert_story_query = 'INSERT INTO stories(words) VALUES(?)'
            cur.execute(insert_story_query, (story_text,))
            story_id = cur.lastrowid
            conn.commit()
            words = story_text.split()  
            fp.close()          
    except Exception as e:
        print("Unable to open the file: " + str(e))
    for word in words:
        if(word in count_of_each_word):
            count_of_each_word[word] += 1
        else:
            count_of_each_word[word] = 1
    for key in count_of_each_word:
        insert_count_query = 'INSERT INTO word_count(words, count ,story_id ) VALUES(?,?,?)'   
        cur.execute(insert_count_query, (key, count_of_each_word[key], story_id))
        conn.commit()
```
